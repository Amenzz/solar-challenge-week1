```python
# replace outlier with median value!!

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("C:/Users/Amenzz/Desktop/kifiya/solar-challenge-week1/data/raw/sierraleone-bumbuna.csv")

# Define sensor columns to analyze
sensor_columns = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# Step 1: Check for missing values
print(" Missing values in key sensor columns:")
display(df[sensor_columns].isna().sum())

# Step 2: Summary statistics
print("\n Summary statistics for sensor columns:")
display(df[sensor_columns].describe())

# Step 3: Visualize outliers using boxplots
plt.figure(figsize=(16, 8))
for i, col in enumerate(sensor_columns):
    plt.subplot(2, 4, i + 1)
    sns.boxplot(x=df[col], color="skyblue")
    plt.title(col)
    plt.tight_layout()
plt.suptitle('Boxplots for Sensor Columns (Outlier Detection)', fontsize=16, y=1.05)
plt.show()

# Step 4: Detect and replace outliers using IQR + median imputation
print("\n Replacing outliers with median values...\n")
for col in sensor_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    median = df[col].median()

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_pct = len(outliers) / len(df) * 100
    print(f"{col}: {len(outliers)} outliers ({outlier_pct:.2f}%)")

    # Replace outliers with median
    df[col] = df[col].apply(lambda x: median if x < lower_bound or x > upper_bound else x)

print("\n Outliers replaced successfully!")

# Optional: Save cleaned dataset
# df.to_csv("data/clean/benin-malanville-cleaned.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore

# Load your dataset
df = pd.read_csv("C:/Users/Amenzz/Desktop/kifiya/solar-challenge-week1/data/raw/sierraleone-bumbuna.csv")

# Define columns to check
sensor_columns = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# Compute Z-scores for selected columns (handling missing values)
z_scores = df[sensor_columns].apply(lambda x: zscore(x, nan_policy='omit'))

# Create a boolean DataFrame: True if |Z| > 3
outlier_flags = (np.abs(z_scores) > 3)

# Show how many outliers are detected in each column
print(" Outliers detected using Z-score (|Z| > 3):")
print(outlier_flags.sum())

# Optionally, get rows with any outlier in selected columns
rows_with_outliers = outlier_flags.any(axis=1)
outlier_df = df[rows_with_outliers]

print(f"\nTotal rows with at least one Z-score outlier: {outlier_df.shape[0]}")

# Optional: Save flagged rows or cleaned version
# outlier_df.to_csv("data/interim/zscore_outliers.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
import os

# Load the data
df = pd.read_csv("C:/Users/Amenzz/Desktop/kifiya/solar-challenge-week1/data/raw/sierraleone-bumbuna.csv")

# Columns to check
key_cols = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# 1. Check for missing values in key columns
print("Missing values before cleaning:")
print(df[key_cols].isnull().sum())

# 2. Compute Z-scores and flag outliers
z_scores = df[key_cols].apply(zscore, nan_policy='omit')
outliers = (z_scores.abs() > 3)
print("\nNumber of outliers in each column:")
print(outliers.sum())

# Optional: Add a flag column for outliers (any key col outlier)
df['OutlierFlag'] = outliers.any(axis=1)

# 3. Handle missing values: impute median for key columns
for col in key_cols:
    median_val = df[col].median()
    df[col] = df[col].fillna(median_value)


# 4. Optionally drop outliers or keep but flagged
# For example, drop all rows with any outlier in key columns
df_clean = df[~df['OutlierFlag']].copy()

# Drop the flag column before export if you want
df_clean.drop(columns=['OutlierFlag'], inplace=True)

# 5. Export cleaned DataFrame
output_dir = "data"
os.makedirs(output_dir, exist_ok=True)  # create if doesn't exist
output_file = os.path.join(output_dir, "sierraleone-bumbuna.csv")
df_clean.to_csv(output_file, index=False)

print(f"\nCleaned data exported to {output_file}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("C:/Users/Amenzz/Desktop/kifiya/solar-challenge-week1/data/raw/sierraleone-bumbuna.csv")

# Convert Timestamp to datetime if not already
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Sort by time just in case
df.sort_values('Timestamp', inplace=True)

# Set Timestamp as index (optional, useful for plotting)
df.set_index('Timestamp', inplace=True)

# Define columns to plot
plot_columns = ['GHI', 'DNI', 'DHI', 'Tamb']  # Tamb = Ambient Temperature

# Time series line plots
plt.figure(figsize=(16, 10))
for i, col in enumerate(plot_columns):
    plt.subplot(len(plot_columns), 1, i + 1)
    sns.lineplot(data=df[col], color="teal")
    plt.title(f'{col} over Time')
    plt.xlabel('Timestamp')
    plt.ylabel(col)
    plt.tight_layout()
plt.suptitle(" Time Series of Key Solar Variables", fontsize=18, y=1.02)
plt.show()

# Extract Month and Hour from Timestamp for analysis
df['Month'] = df.index.month
df['Hour'] = df.index.hour

# Plot monthly patterns
monthly_avg = df.groupby('Month')[plot_columns].mean()
monthly_avg.plot(kind='bar', figsize=(12, 6), colormap='viridis')
plt.title(' Monthly Average of GHI, DNI, DHI, Tamb')
plt.xlabel('Month')
plt.ylabel('Average Value')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Plot hourly patterns
hourly_avg = df.groupby('Hour')[plot_columns].mean()
hourly_avg.plot(kind='line', figsize=(12, 6), marker='o')
plt.title(' Hourly Average of GHI, DNI, DHI, Tamb')
plt.xlabel('Hour of Day')
plt.ylabel('Average Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from windrose import WindroseAxes
import seaborn as sns
import matplotlib.cm as cm  # <-- needed for colormap object


# Load your dataset
df = pd.read_csv("C:/Users/Amenzz/Desktop/kifiya/solar-challenge-week1/data/cleaned/sierraleone-bumbuna.csv")


# ---------------------------
# Wind Rose Plot (WS vs WD)
# ---------------------------
wind_data = df[['WS', 'WD']].dropna()

plt.figure(figsize=(8, 8))
ax = WindroseAxes.from_ax()
ax.bar(
    wind_data['WD'],
    wind_data['WS'],
    opening=0.8,
    edgecolor='white',
    cmap=cm.viridis,  # <-- fixed: use colormap object, not string
    nsector=16
)
ax.set_legend(title="Wind Speed (m/s)")
plt.title(" Wind Rose (Wind Speed vs Direction)")
plt.tight_layout()
plt.show()

# ---------------------------
# Histograms for GHI and WS
# ---------------------------
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# GHI Histogram
sns.histplot(df['GHI'].dropna(), bins=30, kde=True, color='orange', ax=axes[0])
axes[0].set_title(" Histogram of GHI")
axes[0].set_xlabel("GHI (W/m)")
axes[0].set_ylabel("Frequency")

# WS Histogram
sns.histplot(df['WS'].dropna(), bins=30, kde=True, color='skyblue', ax=axes[1])
axes[1].set_title(" Histogram of Wind Speed (WS)")
axes[1].set_xlabel("Wind Speed (m/s)")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()


```

```python
# Temprature Analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("C:/Users/Amenzz/Desktop/kifiya/solar-challenge-week1/data/cleaned/sierraleone-bumbuna.csv")

# Select relevant columns and drop rows with missing values in those columns
df_subset = df[['RH', 'Tamb', 'GHI']].dropna()


# 1. Correlation matrix
print("Correlation matrix:")
print(df_subset.corr())

# 2. Scatter plots with regression lines
plt.figure(figsize=(14, 6))

# RH vs Temperature
plt.subplot(1, 2, 1)
sns.scatterplot(data=df_subset, x='RH', y='Tamb', alpha=0.5)
sns.regplot(data=df_subset, x='RH', y='Tamb', scatter=False, color='red')
plt.title("Relative Humidity vs Tamb")
plt.xlabel("Relative Humidity (%)")
plt.ylabel("Tamb (C)")

# RH vs GHI
plt.subplot(1, 2, 2)
sns.scatterplot(data=df_subset, x='RH', y='GHI', alpha=0.5)
sns.regplot(data=df_subset, x='RH', y='GHI', scatter=False, color='red')
plt.title("Relative Humidity vs Solar Radiation (GHI)")
plt.xlabel("Relative Humidity (%)")
plt.ylabel("GHI (W/m)")

plt.tight_layout()
plt.show()

# 3. Binning RH and visualizing distributions with boxplots
df_subset['RH_bin'] = pd.cut(df_subset['RH'], bins=5)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=df_subset, x='RH_bin', y='Tamb')
plt.title("Tamb distribution by Relative Humidity bins")
plt.xlabel("Relative Humidity bins (%)")
plt.ylabel("Tamb (C)")

plt.subplot(1, 2, 2)
sns.boxplot(data=df_subset, x='RH_bin', y='GHI')
plt.title("Solar Radiation (GHI) distribution by Relative Humidity bins")
plt.xlabel("Relative Humidity bins (%)")
plt.ylabel("GHI (W/m)")

plt.tight_layout()
plt.show()
```
